```python
import cv2
from cvzone.HandTrackingModule import HandDetector
```

```python
cap = cv2.VideoCapture(0)
detector = HandDetector(detectionCon=0.1,maxHands=2)

while cap.isOpened():
    success,frame = cap.read()
    frame = cv2.flip(frame,1)
    frame = cv2.resize(frame,(1700,800))
    # frame = cv2.resize(frame,(800,600))

    hands, frame = detector.findHands(frame)

    if hands:
        hand = hands[0]
        fingers = detector.fingersUp(hand)
        print("Fingers: ",fingers)
        totalFing = fingers.count(1)

        if totalFing == 1 and fingers[0]==1:
            cv2.putText(frame,f"Sign of: A {totalFing}",(200,70),fontFace=cv2.FONT_HERSHEY_DUPLEX,fontScale=1,color=(0,0,255),thickness=2,lineType=4)
        elif totalFing == 4 and fingers[0] == 0:
            cv2.putText(frame,f"Sign of: B {totalFing}",(200,70),fontFace=cv2.FONT_HERSHEY_DUPLEX,fontScale=1,color=(0,0,255),thickness=2,lineType=4)
        elif totalFing == 5:
            cv2.putText(frame,f"Sign of: C {totalFing}",(200,70),fontFace=cv2.FONT_HERSHEY_DUPLEX,fontScale=1,color=(0,0,255),thickness=2,lineType=4)
        elif totalFing == 1 and fingers[1] == 1:
            cv2.putText(frame,f"Sign of: D {totalFing}",(200,70),fontFace=cv2.FONT_HERSHEY_DUPLEX,fontScale=1,color=(0,0,255),thickness=2,lineType=4)
        elif totalFing == 0:
            cv2.putText(frame,f"Sign of: E {totalFing}",(200,70),fontFace=cv2.FONT_HERSHEY_DUPLEX,fontScale=1,color=(0,0,255),thickness=2,lineType=4)
        else:
            cv2.putText(frame,f"Sign of: N/A {totalFing}",(200,70),fontFace=cv2.FONT_HERSHEY_DUPLEX,fontScale=1,color=(0,0,255),thickness=2,lineType=4)



   

    #     cv2.putText(frame,f"No. of fingures up are: {totalFing}",(200,70),fontFace=cv2.FONT_HERSHEY_DUPLEX,fontScale=1,color=(0,0,255),thickness=2,lineType=4)
    cv2.imshow("finger detection", frame)
    if cv2.waitKey(25) & 0xff == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
